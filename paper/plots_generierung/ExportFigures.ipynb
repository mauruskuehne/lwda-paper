```python
import numpy as np
import matplotlib.pyplot as plt
file_perturbation = "Kopie von vggf_xi=10_p=inf_iteration=19.npy"
v = np.load(file_perturbation)
np.min(v),np.max(v),v.shape
```

```python
v[-1,:]; #seltsam!
```

```python
fig = plt.figure(1)
plt.imshow((v[:-1,:]+10)/20, interpolation='none')
ax = plt.gca()
ax.set_xticks([])
plt.xticks([])
plt.yticks([])
ax.set_xticks([])
ax.axis('off')
fig.savefig("vggf.pdf",bbox_inches = 'tight',
    pad_inches = 0)
```
